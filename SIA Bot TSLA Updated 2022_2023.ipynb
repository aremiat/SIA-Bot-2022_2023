```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
analyzer = SentimentIntensityAnalyzer()
```

```python
def analyze_sentiment(tweet):
    sentiment = analyzer.polarity_scores(tweet)
    compound_score = sentiment['compound']
    if compound_score >= 0.5:
        return "Positive"
    elif compound_score <= -0.5:
        return "Negative"
    else:
        return "Neutral"
```

```python
# Data Updated from 2022
data = pd.read_csv(r"C:\Users\alrem\Desktop\Algorithm\Sentiment Analysis\elon_musk_tweets_2009_2024.csv")
data
```

```python
# Probleme car les texts renvoy√©s comportaient des @ et elon musk r√©pondait desfois √† des gens sur TESLA mais qui n'avait rien √† voir avec l'entreprise.
# o√π des gens qui ont tesla dans le pseudo o√π bitcoin exemple @carlaBitcoin beautiful, le sentiment renvoy√© √©tait positif alors que ca n'avait rien √† voir avec quelque chose de positif sur le BTC.
import re

def supprimer_texte_commencant_par_at(texte):
    # Utilisation d'une expression r√©guli√®re pour rechercher les lignes commen√ßant par "@"
    texte_modifie = re.sub(r'@\w+\s*', '', texte)
    return texte_modifie
```

```python
texte_sans_mentions = data["text"].apply(supprimer_texte_commencant_par_at)
print(texte_sans_mentions)
data['Clean_tweets'] = texte_sans_mentions
```

```python
tweets = data["Clean_tweets"]
```

```python
tesla_tweets = data[data['Clean_tweets'].str.lower().str.contains("tesla","TSLA")]
```

```python
tesla_tweets['Sentiment'] = tesla_tweets['Clean_tweets'].apply(analyze_sentiment)
```

```python
print(tesla_tweets[['Clean_tweets', 'Sentiment']])
```

```python
# Initialisation des variables
current_position = "None"  # Au d√©but, aucune position n'est ouverte
cash_balance = 10000.0  # Solde de tr√©sorerie initial
portfolio = {}  # Dictionnaire pour suivre les positions ouvertes



# Logique de trading
for i in range(len(tesla_tweets)):
    tweet_sentiment = tesla_tweets.iloc[i]['Sentiment'] # Avec la fonction iloc, il est possible de r√©cup√©rer toutes les donn√©es des variables √† cette date
    
    # Simulez ici la logique d'achat et de vente en fonction du sentiment
    if tweet_sentiment == "Positive":
        if current_position == "None" and cash_balance > 0:
            # Achat de Tesla
            position_size = 0.2 * cash_balance  # Exemple : investir 20% du solde en TSLA
            cash_balance -= position_size # Cela soustrait le montant de la position achet√©e (position_size) du solde de tr√©sorerie actuel
            portfolio["TSLA"] = position_size # Cela ajoute la position achet√©e dans le dictionnaire portfolio en utilisant le symbole de l'action "TSLA" comme cl√© et la taille de la position (position_size) comme valeur. 
            current_position = "Long"
            print("Achat de Tesla")
    elif tweet_sentiment == "Negative":
        if current_position == "Long" and "TSLA" in portfolio:
            # Vente de Tesla
            position_size = portfolio["TSLA"]
            cash_balance += position_size
            del portfolio["TSLA"]
            current_position = "None"
            print("Vente de Tesla")

# Afficher le solde final du portefeuille
print("Solde final du portefeuille :", cash_balance)
```
